```python
import json
import matplotlib.pyplot as plt
from collections import Counter
```

```python
with open('100_representative.json', 'r') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
# Counter [ham['ham_category'] for ham in data] and plot a bar chart
from collections import Counter
ham_categories = [ham['ham_category'] for ham in data]
ham_categories_counter = Counter(ham_categories)

plt.figure(figsize=(8, 3))

plt.bar(ham_categories_counter.keys(), ham_categories_counter.values())
plt.title('Ham Categories')
plt.show()
```

```python
# Counter [ham['ham_qubits'] for ham in data] and plot a bar chart
ham_qubits = [ham['ham_qubits'] for ham in data if ham['ham_category'] == 'discreteoptimization']
ham_qubits_counter = Counter(ham_qubits)

plt.figure(figsize=(8, 3))
plt.bar(ham_qubits_counter.keys(), ham_qubits_counter.values())
plt.title('Ham Qubits')
plt.show()
```

```python
# Counter [ham['ham_terms'] for ham in data] and plot a bar chart
ham_terms = [ham['ham_terms'] for ham in data if ham['ham_category'] == 'chemistry']
ham_terms_counter = Counter(ham_terms)

plt.figure(figsize=(8, 3))
plt.bar(ham_terms_counter.keys(), ham_terms_counter.values())
plt.title('Ham Terms')
plt.show()
```
